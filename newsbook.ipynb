```python
import pandas as pd

data = pd.read_csv("FA-KES-Dataset.csv",encoding="latin-1 ")
data
```

```python
new_data = data[["article_title","article_content","labels"]]
new_data["news"]=data["article_title"]+" "+data["article_content"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

to_vector = CountVectorizer()

x = to_vector.fit_transform(new_data["news"])
feature_name = to_vector.get_feature_names_out()

y = new_data["labels"]

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)

model = MultinomialNB()
model.fit(x_train,y_train)

user_input = input("Enter the news here")
user_data = to_vector.transform([user_input])

pre = model.predict(user_data)
if pre == 1:
    print("it is likely to real still check more sourses")
else:
    print("fake news")
    
    


```
